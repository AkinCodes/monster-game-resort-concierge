# Monster Resort Concierge - RAGAS Evaluation

This notebook evaluates the RAG system using RAGAS metrics:
- **Faithfulness**: Does the answer follow from the context?
- **Answer Relevancy**: Does the answer address the question?
- **Context Precision**: Is the retrieved context relevant?

We compare three configurations:
1. Basic RAG (dense embeddings only)
2. Hybrid RAG (BM25 + dense)
3. Hybrid + Reranking
